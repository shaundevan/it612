```python
%%writefile code.js
// Part A — filter + map
const data = [5, 12, 3, 20, 11, 9, 30];
const loopA = [];
for (const n of data) {
    if (n > 10) loopA.push(n * 2);
}
const pipelineA = data.filter(n => n > 10).map(n => n * 2);
console.log("A loop:    ", loopA);
console.log("A pipeline:", pipelineA);

// Part B — filter -> map -> reduce
const numbers = [1, 2, 3, 4, 5, 6];
const sumDoubledEvens = numbers
    .filter(n => n % 2 === 0)
    .map(n => n * 2)
    .reduce((acc, n) => acc + n, 0);
console.log("B:", sumDoubledEvens);

// Part C — some, find, forEach
const users = [
    { id: 1, name: "Alice", age: 25 },
    { id: 2, name: "Bob", age: 17 },
    { id: 3, name: "Carol", age: 19 }
];
console.log("C some:", users.some(u => u.age < 18));
console.log("C find:", users.find(u => u.id === 3));
console.log("C forEach:");
users.forEach(u => console.log(u.name));

// Part D — higher-order repeat(n, action)
function repeat(n, action) {
    for (let i = 0; i < n; i++) action(i);
}
console.log("D test 1:");
repeat(3, console.log);
console.log("D test 2:");
repeat(4, i => console.log("Step " + i));
```
